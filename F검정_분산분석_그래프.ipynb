```python
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# 세 가지 약물 투여 후 혈압 감소 수치 (세 그룹의 데이터)
# 약물 A 투여 후 혈압 감소 수치
drug_A = [12, 15, 14, 10, 13]
# 약물 B 투여 후 혈압 감소 수치
drug_B = [18, 20, 17, 19, 22]
# 약물 C 투여 후 혈압 감소 수치
drug_C = [9, 8, 12, 7, 10]

# F 검정 수행 (일원분산분석)
f_stat, p_value = f_oneway(drug_A, drug_B, drug_C)

# 결과 출력
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# 박스 플롯으로 데이터 분포 시각화
plt.figure(figsize=(10, 6))
plt.boxplot([drug_A, drug_B, drug_C], labels=['Drug A', 'Drug B', 'Drug C'])
plt.title("Blood Pressure Reduction by Drug Type")
plt.ylabel("Blood Pressure Reduction")
plt.xlabel("Drug Type")
plt.grid(True)

# ANOVA 결과 주석 추가
plt.text(1.5, max(max(drug_A), max(drug_B), max(drug_C)) + 1, f"F-statistic: {f_stat:.2f}\nP-value: {p_value:.4f}",
         ha='center', fontsize=12, color="red", bbox=dict(facecolor='white', edgecolor='red'))

plt.show()
```
