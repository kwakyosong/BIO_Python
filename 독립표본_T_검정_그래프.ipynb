```python
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# 두 그룹의 데이터
group1 = [88, 78, 64, 92, 57]  # 약물 복용 그룹
group2 = [72, 69, 75, 70, 68]  # 비복용 그룹

# T 검정 수행
t_stat, p_value = ttest_ind(group1, group2)

# 결과 출력
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# 박스 플롯으로 데이터 분포 시각화
plt.figure(figsize=(10, 6))
plt.boxplot([group1, group2], labels=['Drug Group', 'Non-Drug Group'])
plt.title("Comparison of Drug and Non-Drug Groups")
plt.ylabel("Measurement Value")
plt.xlabel("Group")
plt.grid(True)

# T 검정 결과 주석 추가
plt.text(1.5, max(max(group1), max(group2)) + 5,  # y 좌표를 더 위로 설정해 그래프와 겹치지 않게 조정
         f"T-statistic: {t_stat:.2f}\nP-value: {p_value:.4f}",
         ha='center', fontsize=12, color="red", bbox=dict(facecolor='white', edgecolor='red'))

plt.show()
```
